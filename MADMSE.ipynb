```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
OUTPUT_FILE1 = "MAD.png"
OUTPUT_FILE2 = "MSE.png"
```

```python
MADs1 = pd.read_csv("MADs1.csv", names = ["M-Estimator", "MAD"], header = 0)
MADs2 = pd.read_csv("MADs2.csv", names = ["M-Estimator", "MAD"], header = 0)
```

```python
MADs1
```

```python
MADs2
```

```python
xpos = np.arange(len(MADs1))
width = 0.4
offset = width / 2

plt.figure(figsize = (8, 4))
plt.style.use("seaborn-v0_8")

bars1 = plt.bar(x = xpos - offset, height = MADs1.MAD, width = width)
bars2 = plt.bar(x = xpos + offset, height = MADs2.MAD, width = width)
plt.ylabel("Median Absolute Deviation")
plt.xticks(xpos, MADs1["M-Estimator"].str.replace(".", "_"))
plt.yticks([0, 5, 10, 15, 20])
plt.ylim(0, 21)

plt.title("Comparison of MADs for M-Estimators")
plt.legend(["With outliers", "Without outliers"])

plt.bar_label(bars1, fmt='%.2f', fontsize = 9)
plt.bar_label(bars2, fmt='%.2f', fontsize = 9)

plt.tight_layout()
plt.savefig(OUTPUT_FILE1, dpi = 200)
plt.show()
```

```python
MSEs1 = pd.read_csv("MSEs1.csv", names = ["M-Estimator", "MSE"], header = 0)
MSEs2 = pd.read_csv("MSEs2.csv", names = ["M-Estimator", "MSE"], header = 0)
```

```python
xpos = np.arange(len(MADs1))
width = 0.4
offset = width / 2

plt.figure(figsize = (8, 4))
plt.style.use("seaborn-v0_8")

bars1 = plt.bar(x = xpos - offset, height = MSEs1.MSE, width = width)
bars2 = plt.bar(x = xpos + offset, height = MSEs2.MSE, width = width)
plt.ylabel("Mean Squared Error")
plt.xticks(xpos, MADs1["M-Estimator"].str.replace(".", "_"))
plt.yticks([0,1000,2000,3000,4000])
plt.ylim(0, 4500)

plt.suptitle("Comparison of MSEs for M-Estimators")
plt.figlegend(["With outliers", "Without outliers"])

plt.bar_label(bars1, fmt='%.2f', fontsize = 9)
plt.bar_label(bars2, fmt='%.2f', fontsize = 9)

plt.tight_layout()
plt.savefig(OUTPUT_FILE2, dpi = 200)
plt.show()
```
